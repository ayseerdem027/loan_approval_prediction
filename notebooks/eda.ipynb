```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "../data/cleaned/loan_approval_dataset_cleaned.csv"
df = pd.read_csv(file_path)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation")
plt.show()
```

Summary of numerical features:
- when cibil_score increases, loan_status likely becomes 1 (approved)
- loan_term has a very weak negative impact on loan_status

```python
num_columns = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_columns:
    sns.boxplot(x=col, data=df)
    plt.title("Boxplot")
    plt.show()
```

```python
cat_columns = df.select_dtypes(include=['object'])
for col in cat_columns:
    sns.countplot(x = col, hue ='loan_status', data=df)
    plt.title(f"{col} by loan_status")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Summary of categorical features:
- No relationship with loan_status
